```python
data=df.groupby('Gender')['StudyTimeWeekly'].agg(['mean', 'sum', 'count', 'std', 'min', 'max'])
data
```

```python
''' Interpretation

    Mean: On average, the weekly study time is slightly higher for gender 1 (9.835 hours) compared to gender 0 (9.706 hours).
    Sum: The total weekly study time is 12018.878 hours for gender 1 and 11355.726 hours for gender 0.
    Count: There are 1222 observations for gender 1 and 1170 observations for gender 0.
    Std Dev: The standard deviation of weekly study time is slightly higher for gender 1 (5.748) compared to gender 0 (5.554).
    Min: The minimum weekly study time for gender 0 is approximately 0.001 hours, and for gender 1 it is approximately 0.005 hours.
    Max: The maximum weekly study time is approximately 19.979 hours for gender 1 and 19.934 hours for gender 0.
```

```python
data=df.groupby('Absences')['GPA'].agg(['mean', 'sum', 'count', 'std', 'min', 'max'])
data
```

```python
data.columns
```

```python
# Reset index to make the index a column
data = data.reset_index()

# If you want to rename the new column
data = data.rename(columns={'index': 'Absences'})
data
```

```python
# Plotting the data
plt.figure(figsize=(12, 8))

# Plotting mean
plt.plot(data['Absences'], data['mean'], marker='o', label='Mean GPA')

# Plotting sum
plt.plot(data['Absences'], data['sum'], marker='o', label='Sum GPA')

# Plotting count
plt.plot(data['Absences'], data['count'], marker='o', label='Count of Students')

# Adding labels, title, and legend
plt.xlabel('Absences', fontsize=14)
plt.ylabel('Values', fontsize=14)
plt.title('Analysis of GPA Metrics by Absences', fontsize=16)
plt.legend()

# Displaying the plot
plt.tight_layout()
plt.show()
```

```python
#Interpretation
'''
Mean GPA: There is a slight decrease in the mean GPA with more absences, but the change is not drastic. This suggests that while absences have an impact, the average GPA doesn't drop sharply with increasing absences.

Sum GPA: The significant decrease in sum GPA with increasing absences shows that the total academic performance of the group is highly sensitive to absences.

Count of Students: Fewer students have higher absences, which might indicate either better attendance in general or students dropping out or reducing their absences as they realize the negative impact on their academic performance.
```
